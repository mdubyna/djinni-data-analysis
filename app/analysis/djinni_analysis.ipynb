## Import requirements

```python
from collections import Counter

import matplotlib.pyplot as plt
import pandas as pd
```

## Load data

```python
df = pd.read_csv("../filtered_data.csv")
```

## Show most common technologies

```python
df_common = pd.DataFrame(Counter(" ".join(df["description"]).split()).most_common(20))
df_common.set_index(0, inplace=True)
df_common.plot(kind="bar", title="The Popularity of Technologies", xlabel="Technologies", ylabel="Count", rot=60)
```
